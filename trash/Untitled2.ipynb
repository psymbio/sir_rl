```python
import numpy as np
from scipy.optimize import fsolve

# Epidemiological Parameters
Rnot = 2.5  # Total number of transmissions of an infectious (I) person (Source: EC), Atkeson is 3
Tinf = 7  # Duration of infectious period (Source: EC), Atkeson is 18
beta = Rnot / Tinf  # Transmission rate of an infectious (I) person 
gamma = 1 / Tinf  # Exit rate from I state into recovery or death (R)

implicit = 0

T = 400  # Length of simulation
dt = 0.1  # Time step 
Nt = int(T / dt + 1)
time = np.linspace(0, T, Nt)

mu = np.zeros((3, Nt))
S = np.zeros(Nt); I = np.zeros(Nt); R = np.zeros(Nt)
N = np.zeros(Nt); D = np.zeros(Nt)

# Transition matrix
A_t = [None] * Nt

# Initial condition
I0 = 0.35 * 10**(-4)  # March 1st 2020 so that March 31st there are appx. 150,000 cases (50% more than measured)
S0 = 1 - I0
R0 = 0

# SIR MODEL EVENTUAL CASES FOR COMPARISON: solution to log(S) = R_0*(S - 1) or log(1-R) = -R_0*R
fun = lambda x: np.log(x) - Rnot * (x - 1)  # function of x alone
S_st = fsolve(fun, [0.0001, 0.99999])
R_st = 1 - S_st

S_herd = 1. / Rnot
R_herd = (1 - 1. / Rnot)

mu0 = np.array([S0, I0, R0])
mu[:, 0] = mu0

# DO NOTHING
for n in range(Nt - 1):
    S[n] = mu[0, n]
    I[n] = mu[1, n]
    R[n] = mu[2, n]
    Re = Rnot * S
    A_t[n] = np.array([
        [-beta * I[n], beta * I[n], 0],  # from S
        [0, -gamma, gamma],  # from I
        [0, 0, 0]  # from R
    ])

    if implicit == 0:
        mu[:, n + 1] = dt * np.dot(A_t[n].T, mu[:, n]) + mu[:, n]
    else:
        mu[:, n + 1] = np.linalg.solve(np.eye(3) - dt * A_t[n].T, mu[:, n])

mu = mu[:, :-1]

# Convert to percentage terms
mu = 100 * mu

S = mu[0, :]; I = mu[1, :]; R = mu[2, :]; N = S + I + R; D = 100 - N
S_donothing = S; I_donothing = I; R_donothing = R
```
